```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
telecomm_data=pd.read_csv('sample_data/Telcom_Customer_Churn.csv')
telecomm_data.head()
```

```python
telecomm_data.describe().T
```

```python
sns.catplot(y='TotalCharges',kind='box',data=telecomm_data,height=6)
```

```python
sns.catplot(x='Churn',y='TotalCharges',kind='box',data=telecomm_data,height=6)
```

```python
f,ax=plt.subplots(figsize=(10,8))
sns.boxplot(x=telecomm_data['Churn'],y=telecomm_data['TotalCharges'])
```

```python
display(telecomm_data.columns)
```

```python
telecomm_data1=pd.read_csv('sample_data/Telcom_Customer_Churn1.csv')
telecomm_data1.head()
```

```python
telecomm_data1.describe().T
```

```python
f,ax=plt.subplots(figsize=(12,7))
sns.boxplot(x=telecomm_data1['International plan'],y=telecomm_data1['Total intl charge'],palette='rainbow')
```

```python
f,ax=plt.subplots(figsize=(12,7))
sns.boxplot(x=telecomm_data1['International plan'],y=telecomm_data1['Total intl charge'],palette='rainbow',showfliers=False)
```

```python
f,ax=plt.subplots(figsize=(12,7))
sns.boxplot(x='International plan',y='Total intl charge',data=telecomm_data1,hue='Churn',palette='rainbow')
```

```python
#Boxen Plots
sns.catplot(x='International plan',y='Total intl charge',kind='boxen',data=telecomm_data1,hue='Churn',palette='rainbow')
```

```python
#Access level Boxen plots
f,ax=plt.subplots(figsize=(12,7))
sns.boxenplot(x='International plan',y='Total intl charge',data=telecomm_data1,hue='Churn',palette='rainbow')
```

```python
sns.catplot(x='International plan',y='Total intl charge',data=telecomm_data1,kind='box',hue='Churn',col='Voice mail plan',palette='rainbow')
```

```python
sns.catplot(x='International plan',y='Total intl charge',data=telecomm_data1,kind='boxen',hue='Churn',col='Voice mail plan',palette='rainbow')
```
