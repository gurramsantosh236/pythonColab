```python
import seaborn as sns
import pandas as pd
import numpy as np
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
penguins.species.nunique()
```

```python
penguins['species'].value_counts()
```

```python
penguins['island'].nunique()
```

```python
penguins['island'].value_counts()
penguins.info()
```

```python
# scatterplot
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g",hue='sex')
```

```python
# Island colors
island_colors = {"Biscoe": "pink", "Dream": "tab:green", "Torgersen": "tab:orange"}
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g",hue='island',palette=island_colors)
```

```python
# set style function dark grid,whitegrid,dark,white,ticks
sns.set_style("whitegrid")
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g",hue='island',palette=island_colors)
```

```python
# set context paper,notebook,talk,poster
sns.set_context("talk")
sns.scatterplot(data=penguins, x="flipper_length_mm", y="body_mass_g",hue='island',palette=island_colors)
```

```python
# scatter plot for categorical value
sns.set_context("paper")
sns.scatterplot(data=penguins, x="species", y="body_mass_g",hue='island')
```

```python
# strip plot
sns.stripplot(data=penguins, x="species", y="body_mass_g",hue='island',dodge=True)
```

```python
# swarm plots
sns.swarmplot(data=penguins, x="species", y="body_mass_g",hue='island')
```

```python
# hisplot
sns.set_context("notebook")
sns.histplot(data=penguins, x="body_mass_g",hue='sex')
```

```python
# regression plot
sns.regplot(data=penguins, x="body_mass_g", y="flipper_length_mm" , color='green')
```

```python
# line plots
sns.set_style("whitegrid")
sns.lineplot(data=penguins, x="body_mass_g", y="flipper_length_mm",hue='sex')
```

```python
# joint plots
sns.jointplot(data=penguins, x="body_mass_g", y="flipper_length_mm",hue='sex',kind='kde')
```

```python
# Bar plot
sns.barplot(data=penguins, x="species", y="body_mass_g",hue='sex')
```
